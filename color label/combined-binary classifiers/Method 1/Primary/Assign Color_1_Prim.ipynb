```python
#CL1 binary primary
from sklearn.metrics import confusion_matrix,classification_report
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel('test_CL1_prim_binary_xception.xlsx', sheet_name='Sheet1')

id_image = pd.DataFrame(data, columns= ['CL1_Prim','Class'])
print(id_image)
pred = pd.DataFrame(data, columns= ['Predicted_Green','Predicted_Purple','Predicted_Red'])
print(pred)
val = pd.DataFrame(data, columns= ['Softmax_Green','Softmax_Purple','Softmax_Red'])
print(val)

predictor1,predictor2,id1,id2,class1,class2 = [], [], [], [], [], []
for index,row in pred.iterrows():
    print(index)
    print(row)
    colors = []
    vals = []
    if pred.loc[index][0]=='Green':
        color = "Green"
        colors.append(color)
        valss = val.iloc[index,0]
        vals.append(valss)
    if pred.loc[index][1]=='Purple':
        color = "Purple"
        colors.append(color)
        valss = val.iloc[index,1]
        vals.append(valss)
    if pred.loc[index][2]=='Red':
        color = "Red"
        colors.append(color)
        valss = val.iloc[index,2]
        vals.append(valss)
    if pred.loc[index][0]!='Green' and pred.loc[index][1]!='Purple' and pred.loc[index][2]!='Red':
        colors = ["Yellow"]
    
    col = colors
    
    if len(colors)>1:
        colors = ["Inconsistent"]

    print("Predicted Colors1=",colors)
    
    predictor1.extend(list(colors))
    id1.extend([id_image.loc[index][0]])
    class1.extend([id_image.loc[index][1]])

final1 = pd.DataFrame()
final1["id"] = id1
final1["Class"] = class1
final1["Predicted"] = predictor1
final1.to_csv("CL1_BP_1.csv", index=False)
```
