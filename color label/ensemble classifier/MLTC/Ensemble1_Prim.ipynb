```python
#CL1 prim

from sklearn.metrics import confusion_matrix,classification_report
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data_binary = pd.read_csv('CL1_BP_2.csv')
acc_binary = pd.read_csv('testCM_CL1_prim_Xception_Binary_Result2_report.csv')
data_multi = pd.read_excel('test_all_prim_multiclass.xlsx', sheet_name='Sheet3')
#print(data_multi)
acc_multi = pd.read_csv('testCM_CL1_prim_xception_softmax_dropout_freeze 2 bottom layer_report.csv')
#print('Acc_multi',acc_multi)

id_image = pd.DataFrame(data_binary, columns= ['id','Class'])
#print(id_image)
pred1 = pd.DataFrame(data_binary, columns= ['Predicted'])
#print(pred1)
pred2 = pd.DataFrame(data_multi, columns= ['Predicted'])
#print(pred2)
acc_bin = pd.DataFrame(acc_binary.iloc[:, [0,3]])
acc_bin.columns = ['Color', 'Recall']
#print(acc_bin)
acc_mul = pd.DataFrame(acc_multi.iloc[:, [0,3]])
acc_mul.columns = ['Color', 'Recall']
#print(acc_mul)

predictor1,id1,class1,pred_1,pred_2 = [], [], [], [], []
for index,row in pred1.iterrows():
    print(index)
    colors = []
    vals = []
    if pred1.loc[index][0]==pred2.loc[index][0]:
        color = pred1.loc[index][0]      
    else:
        a1 = acc_bin[acc_bin['Color'] == pred1.loc[index][0]]
        a2 = acc_bin[acc_bin['Color'] == pred2.loc[index][0]]
        print(pred1.loc[index][0])
        print(a1)
        print(a2)
        acc_b1 = a1.iloc[0,1]
        acc_b2 = a2.iloc[0,1]
        print(acc_b1)
        print(acc_b2)
        b1 = acc_mul[acc_mul['Color'] == pred2.loc[index][0]]
        b2 = acc_mul[acc_mul['Color'] == pred1.loc[index][0]]
        print(pred2.loc[index][0])
        print(b1)
        print(b2)
        acc_m1 = b1.iloc[0,1]
        acc_m2 = b2.iloc[0,1]
        print(acc_m1)
        print(acc_m2)
        if (acc_b1 > acc_m2) & (acc_m1 < acc_b2):
            color = pred1.loc[index][0]
        else:
            if (acc_b1 > acc_m2) & (acc_m1 > acc_b2):
                if (acc_b1 > acc_m1):
                    color = pred1.loc[index][0]
                else:
                    color = pred2.loc[index][0]
            else:
                if (acc_b1 < acc_m2) & (acc_m1 > acc_b2):
                    color = pred2.loc[index][0]
                else:
                    if (acc_b1 < acc_m2) & (acc_m1 < acc_b2):
                        if (acc_b1 > acc_m1):
                            color = pred1.loc[index][0]
                        else:
                            color = pred2.loc[index][0]

    colors.append(color)
    print("Predicted Colors1=",colors)
    
    pred_1.extend([pred1.loc[index][0]])
    pred_2.extend([pred2.loc[index][0]])
    predictor1.extend(list(colors))
    id1.extend([id_image.loc[index][0]])
    class1.extend([id_image.loc[index][1]])

final1 = pd.DataFrame()
final1["id"] = id1
final1["Class"] = class1
final1["Binary_Predicted"] = pred_1
final1["Multiclass_Predicted"] = pred_2
final1["Ensemble_Predicted"] = predictor1
final1.to_csv("CL1_prim_ensemble2.csv", index=False)
```
